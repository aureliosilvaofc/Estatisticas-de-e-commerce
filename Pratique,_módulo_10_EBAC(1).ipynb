```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('ecommerce_preparados.csv')
df.head()
```

```python
# Informações gerais
df.info()

# Estatísticas descritivas
df.describe()

# Verificando valores nulos
df.isnull().sum()

# Colunas do dataframe
df.columns

# Estilo dos gráficos
sns.set(style="whitegrid")

# Gráfico de Histograma – Distribuição de Idade
plt.figure(figsize=(8,5))
sns.histplot(df['Nota'], kde=True, bins=20)
plt.title('Distribuição das Notas dos Produtos')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()

# Gráfico de Dispersão – Tempo no Site vs Valor da Compra
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Preço', y='Qtd_Vendidos_Cod', hue='Gênero')
plt.title('Preço x Quantidade Vendida')
plt.xlabel('Preço (R$)')
plt.ylabel('Quantidade Vendida')
plt.show()

# Mapa de Calor – Correlação entre Variáveis
plt.figure(figsize=(10,6))
sns.heatmap(df[['Nota', 'N_Avaliações', 'Desconto', 'Preço', 'Qtd_Vendidos_Cod']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor - Correlação entre Variáveis')
plt.show()

# Gráfico de Densidade – Valor da Compra
plt.figure(figsize=(8,5))
sns.kdeplot(df['Preço'], shade=True, color='purple')
plt.title('Densidade de Preço dos Produtos')
plt.xlabel('Preço (R$)')
plt.show()

# Gráfico de Regressão – Tempo no Site vs Valor da Compra
plt.figure(figsize=(8,5))
sns.regplot(data=df, x='Nota', y='Preço', scatter_kws={'alpha':0.3})
plt.title('Regressão: Nota x Preço')
plt.xlabel('Nota')
plt.ylabel('Preço (R$)')
plt.show()
```
